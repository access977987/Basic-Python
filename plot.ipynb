```python
import matplotlib.pyplot as plt
import numpy as np

#生成隨機的data 30個
x = np.random.rand(30)
y = np.random.rand(30)

#製作figure
fig = plt.figure()

#設定axes
ax = fig.add_subplot(1, 1, 1)

#設定散佈圖的顏色，大小等等
ax.scatter(x, y, s=250, alpha=0.4, linewidths=2.5, c='#AAAFFF', edgecolors='blue')

# x, y:X軸、Y軸的數據
# s:散佈圖的點的大小
# c:顏色，這邊也可以寫成color
# marker:散佈圖的點的形狀 (o:圓形、D:菱形、s:四角形、p:五角形、*:星形)
# alpha:散佈圖的點的透明度（可以指定0～1，0:完全透明，1:不透明）
# linewidths:散佈圖的點的邊緣的線的粗度
# edgecolors:散佈圖的點的邊緣的線的顏色

plt.show()
```

```python
left = np.array([1, 2, 3, 4, 5])
height1 = np.array([10, 30,50, 20, 33])
height2 = np.array([21, 31,45, 35, 77])

labels = ['apple', 'banana', 'coconut', 'dragon fruit', 'eggplant']

#選擇要在下面的棒狀圖 blue  誤差3
bar1 = plt.bar(left, height1, color='blue', yerr=3, tick_label=labels)

#選擇要在上面的棒狀圖 red   誤差6
bar2 = plt.bar(left, height2, bottom=height1, yerr=6, color='red', tick_label=labels)
plt.title('Title')
plt.legend((bar1[0], bar2[0]), ('Bar_1', 'Bar_2'))
plt.show()
```

```python
# 製作平均是0，標準差是10的亂數1000個
x = np.random.normal(0, 10, 1000)

# 製作figure
fig = plt.figure()

# 設定ax
ax = fig.add_subplot(1, 1, 1)

# 在axes上畫圖
ax.hist(x, bins=30, density=True)

# 製作平均是0，標準差是10的的機率密度函數
from scipy.stats import norm
X = np.linspace(-50, 50 , 100)
Y = norm.pdf(X, loc=0, scale=10)
ax.plot(X, Y, c='red')

# show
plt.show()
```

Bar plot 的排序
官方文檔搜尋'reorder'

```python
# 一般無排序的bar plot
(
    ggplot(mpg) +
    aes(x='manufacturer') + 
    geom_bar(size=20) + 
    coord_flip() + 
    labs(y='Count', x='Manufacturer', title='Number of Cars by Make')
)
```

```python
# Determine order and create a categorical type
# Note that value_counts() is already sorted
manufacturer_list = mpg['manufacturer'].value_counts().index.tolist() #先依照出現次數做排序並轉換成list

# assign to a new column in the DataFrame
# mpg['manufacturer'] = mpg['manufacturer'].astype('category') # 如果要排序的欄位dtype不是category，要多加這行
mpg = mpg.assign(manufacturer_cat = mpg['manufacturer'].cat.reorder_categories(manufacturer_list)) #把manufacturer的level按照前面的排序，然後新增一欄

(
    ggplot(mpg) + 
    aes(x='manufacturer_cat') + 
    geom_bar(size=20) + 
    coord_flip() + 
    labs(y='Count', x='Manufacturer', title='Number of Cars by Make')
)

```

```python
# 順序相反
manufacturer_list = mpg['manufacturer'].value_counts().index.tolist()[::-1] #先依照出現次數做排序並轉換成list

# assign to a new column in the DataFrame
# mpg['manufacturer'] = mpg['manufacturer'].astype('category') # 如果要排序的欄位dtype不是category，要多加這行
mpg = mpg.assign(manufacturer_cat = mpg['manufacturer'].cat.reorder_categories(manufacturer_list)) #把manufacturer的level按照前面的排序，然後新增一欄

(
    ggplot(mpg) + 
    aes(x='manufacturer_cat') + 
    geom_bar(size=20) + 
    coord_flip() + 
    labs(y='Count', x='Manufacturer', title='Number of Cars by Make')
)
```

Groupby 相關操作
agg (summarise) [max, min, sum, mean, median, std, var, count]
transform
apply
執行速度: agg > transform > apply

assign (mutate)

參考資料: https://zhuanlan.zhihu.com/p/101284491

```python
import numpy as np
import pandas as pd

company = ['A','B','C']
job = ['e', 's', 'q']
data = pd.DataFrame({
    'company':[company[x] for x in np.random.randint(0,len(company),10)],
    'job':[job[x] for x in np.random.randint(0, len(job), 10)],
    'salary':np.random.randint(5,50,10),
    'age':np.random.randint(15,50,10)})

data
```

```python
group = data.groupby("company")
group #顯示是DataFrameGroupBy object
```

```python
data.groupby("company").agg('mean') #注意: groupby的col會變成index，如果要保留col的話，設定as_index=False
```

```python
#針對不同col做不同操作
data.groupby('company').agg({'salary':'median','age':'mean'})
```

```python
#groupby兩個col
data.groupby(['company', 'job'], as_index=False).agg('mean')
```

```python
# transform
# 和agg不同的是transform會根據索引返回function計算的值 (同一個group會賦予相同的值)，如avg_salary這行
data['avg_salary'] = data.groupby('company')['salary'].transform('mean')
data
```

```python
# apply
# 相比agg、transform更靈活，能夠自定義多種function
# 但效率較差，盡量用前兩種

def get_oldest_staff(x):
    df = x.sort_values(by = 'age', ascending = True)
    return df.iloc[-1, :]

data.groupby('company', as_index=False).apply(get_oldest_staff) #as_index如果設為True，groupby的欄位會變成index(row name)，而不會變成一欄
```
